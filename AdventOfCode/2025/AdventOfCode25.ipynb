# Day 6: Trash Compactor

## Description

After helping the Elves in the kitchen, you were taking a break and helping them re-enact a movie scene when you over-enthusiastically jumped into the garbage chute!

A brief fall later, you find yourself in a garbage smasher. Unfortunately, the door's been magnetically sealed.

As you try to find a way out, you are approached by a family of cephalopods! They're pretty sure they can get the door open, but it will take some time. While you wait, they're curious if you can help the youngest cephalopod with her math homework.

Cephalopod math doesn't look that different from normal math. The math worksheet (your puzzle input) consists of a list of problems; each problem has a group of numbers that need to be either added (+) or multiplied (*) together.

However, the problems are arranged a little strangely; they seem to be presented next to each other in a very long horizontal list. For example:

123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  
Each problem's numbers are arranged vertically; at the bottom of the problem is the symbol for the operation that needs to be performed. Problems are separated by a full column of only spaces. The left/right alignment of numbers within each problem can be ignored.

So, this worksheet contains four problems:

123 * 45 * 6 = 33210
328 + 64 + 98 = 490
51 * 387 * 215 = 4243455
64 + 23 + 314 = 401
To check their work, cephalopod students are given the grand total of adding together all of the answers to the individual problems. In this worksheet, the grand total is 33210 + 490 + 4243455 + 401 = 4277556.

Of course, the actual worksheet is much wider. You'll need to make sure to unroll it completely so that you can read the problems clearly.

Solve the problems on the math worksheet. What is the grand total found by adding together all of the answers to the individual problems?

```python
## Part 1

import operator
from functools import reduce

maths_worksheet = [list(w.split()) for w in input_maths_worksheet.split('\n')]
worksheet_numbers = [[int(x) for x in row] for row in maths_worksheet[:-1]]
worksheet_operators = maths_worksheet[-1]
worksheet_results = []

#print(maths_worksheet)
#print(worksheet_numbers)
#print(worksheet_operators)

ops = {
    '+': operator.add,
    '-': operator.sub,
    '*': operator.mul,
    '/': operator.truediv,
}

for i_column, operator in enumerate(worksheet_operators):
    problem_numbers = [c[i_column] for c in worksheet_numbers]
    worksheet_results.append(reduce(ops[operator], problem_numbers))

print(worksheet_results)
print('Sum:',sum(worksheet_results))
```

```python
## Part 2

import operator
import re
from functools import reduce

# Seperate and store as strings to maintain column format
maths_worksheet_strings = input_maths_worksheet.split('\n')
worksheet_numbers_strings = maths_worksheet_strings[:-1]
worksheet_operators_strings = maths_worksheet_strings[-1]
worksheet_operators = worksheet_operators_strings.split() # to use as operators later

# Determine the number of digits per worksheet problem from the operator being left aligned
spaces = re.findall(r'\s+', worksheet_operators_strings)
digits_per_problem = [len(space) for space in spaces]
digits_per_problem[-1] = digits_per_problem[-1]+1 # to account for no space dilimiter in the last column

# Go down columns to get strings and then numbers for operating
column_strings = [''.join(col) for col in zip(*worksheet_numbers_strings)] # go down each column
column_strings = [x for x in column_strings if x.strip() != ''] # remove columns where all spaces i.e. delimeter columns
column_numbers = [int(c) for c in column_strings] # convert strings to numbers

# Group columns which are part of the same operation into the problems using the digits per worksheet problem
column_numbers_grouped = []
start = 0
for d in digits_per_problem:
    column_numbers_grouped.append(column_numbers[start:start + d])
    start += d

# Perform problem operation
worksheet_results = []
for i_column, operator in enumerate(worksheet_operators):
    worksheet_results.append(reduce(ops[operator], column_numbers_grouped[i_column]))

print('Sum:',sum(worksheet_results))
```
